## **1. Pengantar BAB 3**

Bab 3 membahas dasar implementasi fisik basis data menggunakan **SQL  bagian Data Definition Language (DDL)**. Setelah mahasiswa memahami ERD dan skema relasi (BAB 12), tahap berikutnya adalah membangun struktur database nyata menggunakan perintah DDL pada DBMS seperti MySQL.

DDL berfungsi **membentuk kerangka** database meliputi:

* membuat database,
* membuat tabel,
* mengatur tipe data,
* menentukan primary key & foreign key,
* membuat hubungan tabel,
* dan menetapkan aturan integritas data (constraints).

Materi pada bab ini menjadi pondasi untuk operasi data pada bab-bab berikutnya.

---

## **2. Konsep Dasar DDL**

DDL (Data Definition Language) adalah bagian dari SQL yang digunakan untuk mendefinisikan struktur database. Perintah dalam DDL bersifat **mengubah arsitektur** database, bukan datanya.

Beberapa perintah utama DDL:

1. **CREATE**  membuat database atau tabel
2. **ALTER**  mengubah struktur tabel
3. **DROP**  menghapus tabel atau database
4. **TRUNCATE**  menghapus seluruh isi tabel tanpa menghapus tabelnya

Pada Bab 3, fokus utama adalah **CREATE TABLE** dan aturan pembentukan struktur.

---

## **3. Pembuatan Database**

Pembuatan database dilakukan sebelum membuat tabel. Sintaks umum:

```
CREATE DATABASE nama_database;
USE nama_database;
```

Database dapat dihapus dengan:

```
DROP DATABASE nama_database;
```

Materi ini menekankan pentingnya konsistensi penamaan dan identitas database (misal berdasarkan NIM atau nama praktikan).

---

## **4. Pembuatan Tabel Menggunakan CREATE TABLE**

Inti dari BAB 3 adalah memahami pembuatan tabel secara lengkap sesuai **skema relasi** hasil BAB 1.

Struktur dasar perintah:

```
CREATE TABLE nama_tabel (
    kolom1 tipe_data constraint,
    kolom2 tipe_data,
    ...
    PRIMARY KEY (...),
    FOREIGN KEY (...) REFERENCES tabel_lain(kolom)
) ENGINE=InnoDB;
```

Hal-hal penting dalam pembuatan tabel:

### **4.1 Tipe Data**

Beberapa tipe data yang sering dipakai:

* **CHAR, VARCHAR**  teks
* **INT, BIGINT**  bilangan
* **DECIMAL**  bilangan presisi
* **DATE, DATETIME**  tanggal/waktu
* **BOOLEAN**  benar/salah

### **4.2 Aturan Penamaan Kolom**

* nama kolom harus jelas dan mewakili isi
* menggunakan huruf kecil atau snake_case
* tidak boleh memakai spasi

### **4.3 Engine Tabel: InnoDB**

InnoDB digunakan karena mendukung:

* foreign key
* integritas referensial
* transaksi
* relasi antar tabel

Materi modul menekankan bahwa **jika tabel memakai relasi, wajib InnoDB**.

---

## **5. Constraints (Aturan Integritas Data)**

Constraint digunakan untuk memastikan data tetap konsisten.
BAB 3 membahas constraint berikut:

### **5.1 PRIMARY KEY**

* penanda unik untuk setiap baris
* tidak boleh null
* dapat berupa 1 kolom atau gabungan
* disarankan menggunakan kode unik

### **5.2 FOREIGN KEY**

* penghubung antara tabel satu dengan lainnya
* menjaga integritas referensial
* memastikan data parent sudah ada sebelum child dibuat

Format:

```
FOREIGN KEY (id_kategori)
REFERENCES kategori(id_kategori)
```

### **5.3 UNIQUE**

* mencegah nilai duplikat

### **5.4 NOT NULL**

* kolom harus diisi

### **5.5 CHECK**

* membatasi nilai tertentu (meski pada MySQL tidak semua versi mendukung)

### **5.6 DEFAULT**

* memberi nilai otomatis saat tidak ada input

Contoh:

```
status VARCHAR(10) DEFAULT 'Aktif'
```

---

## **6. AUTO_INCREMENT**

Digunakan untuk membuat penomoran otomatis, biasanya untuk primary key berbasis angka.

Contoh:

```
id INT AUTO_INCREMENT PRIMARY KEY
```

Materi menjelaskan kapan perlu Auto Increment dan kapan tidak (misal untuk NIM atau kode obat yang bukan berupa angka berurutan).

---

## **7. Relasi Antar Tabel**

Setelah tabel dasar dibuat, mahasiswa diminta membuat hubungan tabel sesuai ERD menggunakan:

* **PRIMARY KEY** pada tabel induk
* **FOREIGN KEY** pada tabel anak

Contoh relasi:

```
CREATE TABLE transaksi (
    id_transaksi INT AUTO_INCREMENT PRIMARY KEY,
    id_pasien INT,
    FOREIGN KEY (id_pasien) REFERENCES pasien(id_pasien)
);
```

BAB 3 mengulang kembali konsep relasi:

* One-to-One
* One-to-Many (yang paling sering)
* Many-to-Many (memerlukan tabel penghubung)

Materi menuntut mahasiswa memahami *makna relasi* sebelum menuliskan foreign key.

# **SOAL 1**

1. Pak Rizq adalah seorang profesor yang ingin membuat sistem sederhana untuk mencatat nilai tugas siswa. Selama ini ia mencatat di buku saja sehingga sering salah memasukkan nilai. Ia ingin membuat database untuk membantu proses pencatatan.

   Pak Rizq membutuhkan fitur sebagai berikut:

   1. Data siswa terdiri dari nama, kelas, nomor induk
   2. Data tugas terdiri dari judul tugas, tanggal diberikan, dan tanggal deadline
   3. Dapat menyimpan nilai siswa

```sql
CREATE DATABASE db_pak_rizq;
USE db_pak_rizq;

CREATE TABLE Siswa (
    nomor_induk INT PRIMARY KEY,
    nama_siswa VARCHAR(100) NOT NULL,
    kelas VARCHAR(10) NOT NULL
);

CREATE TABLE Tugas (
    id_tugas INTEGER PRIMARY KEY AUTO_INCREMENT,
    judul_tugas VARCHAR(100) NOT NULL,
    tanggal_diberikan DATE NOT NULL,
    tanggal_deadline DATE NOT NULL
);

CREATE TABLE Nilai (
    id_nilai INTEGER PRIMARY KEY AUTO_INCREMENT,
    nomor_induk INT NOT NULL,
    id_tugas INT NOT NULL,
    nilai INT NOT NULL,
    FOREIGN KEY (nomor_induk) REFERENCES Siswa(nomor_induk),
    FOREIGN KEY (id_tugas) REFERENCES Tugas(id_tugas)
);


```

2. Pak Ucup adalah seorang tukang jualan somay di depan sekolah. Ia ingin membuat aplikasi web sederhana agar usahanya lebih rapi. Selama ini ia mencatat di kertas kecil dan berjualan secara offline.

```sql  
CREATE DATABASE db_pak_ucup;
USE db_pak_ucup;

CREATE TABLE Menu (
    id_menu INT PRIMARY KEY AUTO_INCREMENT,
    nama_produk VARCHAR(100) NOT NULL,
    harga DECIMAL(10, 2) NOT NULL
);

CREATE TABLE Pesanan (
    id_pesanan INT PRIMARY KEY AUTO_INCREMENT,
    tanggal_pesanan DATETIME NOT NULL,
    nama_pelanggan VARCHAR(100),
    status_pesanan VARCHAR(50) NOT NULL DEFAULT 'Baru'
);

CREATE TABLE Detail_Pesanan (
    id_detail INT PRIMARY KEY AUTO_INCREMENT,
    id_pesanan INT NOT NULL,
    id_menu INT NOT NULL,
    kuantitas INT NOT NULL,
    harga_saat_pesan DECIMAL(10, 2) NOT NULL,
    FOREIGN KEY (id_pesanan) REFERENCES Pesanan(id_pesanan),
    FOREIGN KEY (id_menu) REFERENCES Menu(id_menu),
    UNIQUE (id_pesanan, id_menu)
);



```

---

## **9. Kesimpulan BAB 3**

* Bab 3 adalah tahap awal implementasi fisik database.
* Fokus utama adalah pembuatan struktur tabel dengan DDL.
* Mahasiswa mempelajari CREATE TABLE, tipe data, dan semua jenis constraint.
* InnoDB digunakan untuk mendukung relasi antar tabel.

---
